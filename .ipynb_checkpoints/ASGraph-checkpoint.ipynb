```python
import sys
import matplotlib.pyplot as plt
from pprint import pprint
import networkx as nx


graph_str = ""
i = 0
with open('cycles-aslinks.txt') as fp:
    for l in fp:
        if l.startswith('D'):# or l.startswith('I'):
            i += 1
            if i > 10000:
                break
            l = l.strip()
            l_ls = l.split('\t')
            tuple_str = '{0} {1}\n'.format(l_ls[1], l_ls[2])
            graph_str = '{0} {1}'.format(graph_str, tuple_str)
```

```python
with open('cycles_parsed.txt','w') as fp:
    fp.write(graph_str[:-1])
```

```python
H = nx.read_edgelist(path="cycles_parsed.txt")
nx.draw(H, node_size=0.2)
plt.show()
```

```python
plt.savefig('graph.png')
```
